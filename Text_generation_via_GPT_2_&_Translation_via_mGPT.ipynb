```python
import os
import torch



if torch.cuda.is_available():
    device = "cuda"
    os.environ["CUDA_VISIBLE_DEVICES"] = "0"
    os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
else:
    device = "cpu"

print(device)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel


# создадим инстанс GPT2Tokenizer, инициализировав его
# моделью gpt2-large с помощью метода from_pretrained
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
# создадим инстанс GPT2LMHeadModel, инициализировав его
# моделью gpt2-large с помощью метода from_pretrained
model = GPT2LMHeadModel.from_pretrained('gpt2-large')
# сразу перенесем модель на девайс с помощью метода to
device="cpu"
model.to(device)
```

```python
torch.cuda.is_available()
```

```python
from transformers import pipeline


# Создаем pipeline обработки текста для задачи text-generation,
# который получает на вход модель, токенизатор и индекс доступного девайса
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, device='cpu')
```

```python
sentence = 'I want to buy this'
generated = pipe(sentence)
print(generated[0]["generated_text"])
```

Перевод используя модель MGPT

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
# создадим инстанс GPT2Tokenizer, инициализировав его
# моделью sberbank-ai/mGPT с помощью метода from_pretrained
tokenizer = GPT2Tokenizer.from_pretrained('sberbank-ai/mGPT')
# создадим инстанс GPT2LMHeadModel, инициализировав его
# моделью sberbank-ai/mGPT с помощью метода from_pretrained
model = GPT2LMHeadModel.from_pretrained('sberbank-ai/mGPT')
# сразу перенесем модель на девайс с помощью метода to
device = "cpu"
model.to(device)
```

```python
from transformers import pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, device='cpu')
```

```python
sentence = ''' \
Translate Русский to English: \

Русский: Сегодня отличный матч! Chinese: '''

generated = pipe(sentence, max_length=40)
print(generated[0]["generated_text"])
```
